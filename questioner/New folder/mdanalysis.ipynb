```python
from markitdown import MarkItDown

md = MarkItDown()
results = []

# Convert each Excel file and store the results
files = ["r1.pdf", "r2.pdf", "r3.pdf", "r4.pdf"]
for file in files:
    result = md.convert(file)
    results.append(result)
```

```python
# Print the results
for i, result in enumerate(results, start=1):
    print(f"Result for file {i}:")
    print(result)
    print("\n" + "="*40 + "\n")
```

```python
# save into md file
with open("results.md", "w") as f:
    for i, result in enumerate(results, start=1):
        f.write(f"Result for file {i}:\n")
        f.write(result.text_content + "\n")
        f.write("\n" + "="*40 + "\n")
```
