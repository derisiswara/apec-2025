## Analysis of APEC 2025 Survey Data

```r
# Load necessary libraries
library(tidyverse)
library(ggplot2)
library(scales)
library(gridExtra)
library(RColorBrewer)
library(knitr)
library(readxl)
library(patchwork) # For combining plots

# Import data
dataactor <- read_excel("data/pelaku.xlsx")

# Basic data overview
cat("Number of respondents:", nrow(dataactor), "\n")
cat("Number of questions:", ncol(dataactor) - 12, "\n")  # Subtracting demographic columns

# Data preprocessing for analysis
# Create standardized province names
dataactor$Province <- recode(dataactor$Province,
                            "Dki Jakarta" = "Jakarta",
                            "DKI Jakarta" = "Jakarta",
                            "DIY" = "Yogyakarta",
                            "Jawa Tengah" = "Central Java",
                            "Jawa Barat" = "West Java")

# Create a standardized education category
dataactor$Edu <- factor(
  case_when(
    dataactor$`Level of Education` %in% c("Primary School", "Junior HS", "Senior HS") ~ "Basic Education",
    dataactor$`Level of Education` %in% c("Vocational School", "3y Diploma") ~ "Vocational",
    dataactor$`Level of Education` %in% c("Bachelor", "Bachelor Adm Negara", "Master Degre") ~ "College Degree",
    TRUE ~ "Other"
  ),
  levels = c("Basic Education", "Vocational", "College Degree"),
  ordered = TRUE
)
```

## Compliance Gaps

```r
regulatory_questions <- c(
	"3. Does your business have a business license/legality?",
	"4. Does the product being sold have a distribution permit?",
	"73. Are there records and documentation regarding the expiration date of raw materials, incoming orders, pest control, cleaning schedules for facilities and business premises?",
	"74. Are there any records regarding worker health checks, training schedules, worker leave schedules?"
)

# Function to convert Yes/No to numeric
convert_to_numeric <- function(x) {
	case_when(
		str_detect(tolower(x), "yes|ya") ~ 1,
		str_detect(tolower(x), "no|tidak") ~ 0,
		TRUE ~ NA_real_
	)
}

# Calculate compliance scores
compliance_data <- dataactor %>%
	select(all_of(regulatory_questions)) %>%
	mutate_all(convert_to_numeric) %>%
	rowwise() %>%
	mutate(compliance_score = mean(c_across(everything()), na.rm = TRUE)) %>%
	ungroup()

# BBC style colors
bbc_blue <- "#1380A1"
bbc_light_blue <- "#BEDFED"
bbc_red <- "#990000"
bbc_gray <- "#404040"
bbc_light_gray <- "#DDDDDD"

# Plot 1: Compliance Distribution
options(repr.plot.width = 12, repr.plot.height = 8)
p1 <- compliance_data %>%
	ggplot(aes(x = compliance_score)) +
	geom_histogram(bins = 5, fill = bbc_blue, color = "white", alpha = 0.9) +
	scale_x_continuous(labels = percent_format(), 
										 limits = c(0, 1),
										 breaks = seq(0, 1, 0.25)) +
	labs(
		title = "Regulatory Compliance Gap",
		subtitle = "Only 48% of vendors meet basic requirements",
		x = NULL,
		y = NULL) +
	bbplot::bbc_style() +
	annotate(
    'text',
    x = 0.8,
    y = 13,
    label = paste0(round(mean(compliance_data$compliance_score, na.rm = TRUE) * 100), "% average \ncompliance"),
    fontface = 'bold', 
    size = 10
  ) +
  annotate(
    'curve',
    x = 0.8, # Play around with the coordinates until you're satisfied
    y = 11,
    yend = 12,
    xend = 0.5,
    linewidth = 1,
    curvature = -0.5,
    arrow = arrow(length = unit(0.5, 'cm'))
  )
p1
```
